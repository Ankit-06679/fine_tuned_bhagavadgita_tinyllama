```python
import pandas as pd

df = pd.read_csv("/content/Bhagwad_Gita.csv")  # adjust filename if different
df.head()


```

```python
print(df.columns)
print(df.head(1).T)
```

```python
import json

instruction_data = []

for _, row in df.iterrows():
    instruction = f"Explain Bhagavad Gita Chapter {row['Chapter']} Verse {row['Verse']}"
    output = f"Sanskrit: {row['Shloka']}\nTranslation: {row['EngMeaning']}"

    instruction_data.append({
        "instruction": instruction,
        "input": "",
        "output": output
    })

# Save to file
with open("gita_instructions.json", "w") as f:
    json.dump(instruction_data, f, indent=2)

print(f"✅ Instruction-style dataset created with {len(instruction_data)} records.")
```

```python
from datasets import load_dataset

dataset = load_dataset("json", data_files="gita_instructions.json")
dataset = dataset["train"].train_test_split(test_size=0.1)  # 90% train, 10% test
dataset["train"][0]
```

```python
from transformers import AutoTokenizer

model_id = "TinyLlama/TinyLlama-1.1B-Chat-v1.0"

tokenizer = AutoTokenizer.from_pretrained(model_id, trust_remote_code=True)
tokenizer.pad_token = tokenizer.eos_token  # Set padding to EOS
```

```python
def format_prompt(example):
    return f"### Instruction:\n{example['instruction']}\n\n### Response:\n{example['output']}"

def tokenize(example):
    prompt = format_prompt(example)
    result = tokenizer(prompt, truncation=True, max_length=512, padding="max_length")
    result["labels"] = result["input_ids"].copy()
    return result

tokenized_dataset = dataset.map(tokenize, remove_columns=dataset["train"].column_names)
```

```python
from transformers import AutoModelForCausalLM
import torch

model = AutoModelForCausalLM.from_pretrained(
    model_id,
    load_in_8bit=True,
    device_map="auto",
    torch_dtype=torch.float16
)

from peft import LoraConfig, get_peft_model, TaskType

lora_config = LoraConfig(
    r=8,
    lora_alpha=16,
    target_modules=["q_proj", "v_proj"],
    lora_dropout=0.1,
    bias="none",
    task_type=TaskType.CAUSAL_LM
)

model = get_peft_model(model, lora_config)
model.print_trainable_parameters()

from transformers import TrainingArguments, Trainer, DataCollatorForLanguageModeling

training_args = TrainingArguments(
    output_dir="gita-tinyllama",
    per_device_train_batch_size=4,
    per_device_eval_batch_size=4,
    gradient_accumulation_steps=4,
    num_train_epochs=3,
    learning_rate=2e-4,
    logging_steps=10,
    fp16=True,
    save_total_limit=2,
    save_steps=50,  # instead of save_strategy
    eval_steps=50,  # instead of evaluation_strategy
)

data_collator = DataCollatorForLanguageModeling(tokenizer, mlm=False)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset["train"],
    eval_dataset=tokenized_dataset["test"],
    tokenizer=tokenizer,
    data_collator=data_collator
)

trainer.train()
```

```python
trainer.save_model("gita-tinyllama-lora")
tokenizer.save_pretrained("gita-tinyllama-lora")
```

```python
from transformers import pipeline

pipe = pipeline("text-generation", model=model, tokenizer=tokenizer)

prompt = """### Instruction:
Explain Bhagavad Gita Chapter 2 Verse 47

### Response:"""

response = pipe(prompt, max_new_tokens=256, temperature=0.7, top_p=0.9, repetition_penalty=1.2)[0]['generated_text']
print(response)

```

```python
import nest_asyncio
import threading
import uvicorn
from fastapi import FastAPI, Request
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
from peft import PeftModel

app = FastAPI()
nest_asyncio.apply()

@app.post("/generate")
async def generate(request: Request):
    data = await request.json()
    prompt = f"### Instruction:\n{data['prompt']}\n\n### Response:"
    result = pipe(prompt, max_new_tokens=256, temperature=0.7, top_p=0.9)[0]['generated_text']
    return {"response": result.split("### Response:")[-1].strip()}

# Load model
base_model_id = "TinyLlama/TinyLlama-1.1B-Chat-v1.0"
lora_model_path = "gita-tinyllama-lora"

tokenizer = AutoTokenizer.from_pretrained(base_model_id)
base_model = AutoModelForCausalLM.from_pretrained(base_model_id, device_map="auto", load_in_8bit=True)
model = PeftModel.from_pretrained(base_model, lora_model_path)

pipe = pipeline("text-generation", model=model, tokenizer=tokenizer)

# Run FastAPI server
def run():
    uvicorn.run(app, host="0.0.0.0", port=8000)

thread = threading.Thread(target=run)
thread.start()
```
